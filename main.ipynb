```python
import glob
import time
import numpy as np
from zxcvbn import zxcvbn

filepath = './passwords/'
filename = '*.txt'
list_txt = glob.glob(filepath+filename)
for txt_name in list_txt:
    t = time.time()
    print(txt_name)
    data = np.loadtxt(txt_name, dtype=str)
    print(data.shape)
    guess_log10 = np.zeros(data.shape)
    for index in range(data.shape[0]):
        result = zxcvbn(data[index])
        guess_log10[index] = result['guesses_log10']
    csvname = txt_name[0:-4] + '_guesslog10.csv'
    np.savetxt(csvname, guess_log10, delimiter=",")
    elapsed = time.time() - t
    print("Finished in second: ", elapsed)
```
